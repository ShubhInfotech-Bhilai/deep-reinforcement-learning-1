```python
import matplotlib.pyplot as plt
import os
import cv2
```

```python
%matplotlib inline
```

```python
os.listdir()
```

```python
img=plt.imread('Images (4).jpg')
```

```python
plt.figure(figsize=(10,10))
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(10,10))
plt.xticks([])
plt.yticks([])
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
import numpy as np
```

```python
sobel_y = np.array([[0,-1,0],
                   [-1,4,-1],
                   [0,-1,0]])

filtererd_image = cv2.filter2D(gray,-1,sobel_y)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(filtererd_image,cmap='gray',interpolation='nearest')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#Vizualizing 100 random filters
```

```python
x = np.random.randint(low=-10,high=10,size=(200,4,4))
a = []
for i in range(x.shape[0]):
    a.append(cv2.filter2D(gray,-1,x[i]))
a = np.squeeze(np.array(a))
plt.figure(figsize=(100,100))
for i in range(a.shape[0]):
    plt.subplot(50,40,i+1)
    plt.imshow(a[i],cmap='gray',interpolation='nearest')
    plt.xticks([])
    plt.yticks([])
plt.show()
```
