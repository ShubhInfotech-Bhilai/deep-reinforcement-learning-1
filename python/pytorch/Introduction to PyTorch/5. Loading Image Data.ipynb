```python
import torch
%matplotlib inline
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
import numpy as np
from torchvision import datasets, transforms
```

```python
transform = transforms.Compose([transforms.RandomRotation(30),
                                transforms.RandomResizedCrop(100),
                                transforms.RandomHorizontalFlip(),
                               transforms.CenterCrop(224),
                               transforms.Resize(255),
                               transforms.ToTensor()])
```

```python
data = datasets.ImageFolder('imgs/',transform=transform)
```

```python
dataloader = torch.utils.data.DataLoader(data,batch_size=1,shuffle=True)
```

```python
image, label = next(iter(dataloader))
```

```python
image.shape
```

```python
plt.figure(figsize=(10,10))
for i in range(len(dataloader)):
    image, label = next(iter(dataloader))
    plt.subplot(4,4,i+1)
    plt.imshow(image[0].reshape(image.shape[2],image.shape[3],image.shape[1]),interpolation='nearest')
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(label.numpy())
```
